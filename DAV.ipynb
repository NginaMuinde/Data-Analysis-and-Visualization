```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for seaborn
sns.set(style="whitegrid")
```

```python
#Load and Explore the Dataset
# Load the dataset
try:
    df = pd.read_csv('dm_office_sales.csv')
except FileNotFoundError:
    print("Error: The file 'dm_office_sales.csv' was not found.")
```

```python
# Display the first few rows of the dataset
print(df.head())
```

```python
# Explore the structure of the dataset
print(df.info())
print(df.isnull().sum())
```

```python
#Clean the dataset
#Fill missing values
df.fillna(value=0, inplace=True)
```

```python
# Compute basic statistics
print(df.describe())
```

```python
#Check colums of dataset
print(df.columns)
```

```python
# Perform groupings and compute means
grouped_data = df.groupby('division')['sales'].mean()
print(grouped_data)
```

```python
# Group by 'division' and compute mean sales
grouped_data = df.groupby('division')['sales'].mean()

# Identify highest sales category
print("Highest sales category:", grouped_data.idxmax(), "with sales:", grouped_data.max())
```

```python
# 2. Bar chart showing average sales per category
plt.figure(figsize=(10, 6))
grouped_data.plot(kind='bar')
plt.title('Average Sales per Category')
plt.xlabel('Category')
plt.ylabel('Average Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Histogram of sales distribution
plt.figure(figsize=(10, 6))
df['sales'].hist(bins=30)
plt.title('sales Distribution')
plt.xlabel('sales')
plt.ylabel('Frequency')
plt.show()
```

```python
# 4. Scatter plot of training level vs sales
plt.figure(figsize=(10, 6))
plt.scatter(df['training level'], df['sales'], alpha=0.5)
plt.title('sales vs training level')
plt.xlabel('training level')
plt.ylabel('sales')
plt.show()
```
